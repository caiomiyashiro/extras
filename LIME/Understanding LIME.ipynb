```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('breast-cancer-wisconsin.csv', index_col=0)
df = df.iloc[:,:-1]
df.head()
```

```python
df_train = df.drop('diagnosis', axis=1, inplace=False)
target = [1 if elem == 'M' else 0 for elem in df['diagnosis']]
```

```python
from tpot import TPOTClassifier

X_train, X_test, y_train, y_test = train_test_split(df_train, target,
                                                    train_size=0.75, test_size=0.25)

tpot = TPOTClassifier(generations=5, population_size=20, verbosity=2)
tpot.fit(X_train, y_train)
print(tpot.score(X_test, y_test))
```
